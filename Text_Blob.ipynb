```
import pandas as pd
```

```
data = pd.read_csv("labeledTrainData.tsv",header=0, delimiter="\t", quoting=3)
```

```
data.head()
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( data['review'], data['sentiment'], test_size=0.33, random_state=42)
```

```
from textblob import TextBlob
```

```
def text_blob_sentiment(text):
    analysis = TextBlob(text)
    return 1 if (analysis.sentiment.polarity > 0.1) else 0
```

```
pred_y = [text_blob_sentiment(text) for text in X_train]
```

```
from sklearn.metrics import accuracy_score
```

```
print(accuracy_score(y_train, pred_y))
```
